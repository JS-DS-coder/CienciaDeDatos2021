# Evaluacion con Polyfit


```python
def EvaluacionPolyfyt(X,Y,grado,X_seq,nombre):
  degree=grado
  polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
  polyreg.fit(X,Y)
  plt.figure()
  plt.scatter(X,Y)
  plt.plot(X_seq,polyreg.predict(X_seq),color="red")
  plt.title("Regresion Polinomial columna " + nombre + "  con grado "+str(degree))
  plt.show()
```
