MONOCROCMO

```python
## Dimensiones de la imagen
alto = img.shape[0]
ancho = img.shape[1]
```

```python
imgSalida = np.zeros((alto,ancho,3))
```

```python
for i in range(0,alto):
  for j in range (0,ancho):
    pixel = img[i,j]
    blue = pixel[2]
    green = pixel[1]
    red = pixel [0]
    imgSalida[i,j] = [0,green,0]

#esto me va a devolver un tensor, pero no es una imagen, se debe de convertir este arreglo a una imagen.
```

```python
#la bajamos a disco para convertir el arreglo a imagen
cv.imwrite("Imagen_salida.jpg",imgSalida)
imgTemp = cv.imread("Imagen_salida.jpg")
imgTemp = cv.cvtColor(imgTemp,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(imgTemp)
```

**ESCALA DE GRISES**

```python
gray_green = np.zeros((alto,ancho,1))
gray_blue = np.zeros((alto,ancho,1))
gray_red = np.zeros((alto,ancho,1))

gray_aritmentica = np.zeros((alto,ancho,1))
gray_ponderada = np.zeros((alto,ancho,1))
```

```python
for i in range(0,alto):
  for j in range (0,ancho):
    pixel = img[i,j]
    gray_blue[i,j] = pixel[2]
    gray_green[i,j] = pixel[1]
    gray_red[i,j] = pixel [0]

    gray_aritmentica[i,j] = np.mean(pixel)
    gray_ponderada[i,j] = int(0.299*pixel[2] + 0.587*pixel[1] + 0.114*pixel[0])
    imgSalida[i,j] = [0,green,0]
```

```python
def write_and_show (strImg,arrImg):
  cv.imwrite(strImg,arrImg)
  imgTemp = cv.imread(strImg)
  imgTemp = cv.cvtColor(imgTemp,cv.COLOR_BGR2RGB)
  plt.imshow(imgTemp)
```

```python
write_and_show("grayrojo.jpg",gray_red)

```

```python
write_and_show("grayblue.jpg",gray_blue)
```

```python

write_and_show("graygreen.jpg",gray_green)
```

```python
write_and_show("gray_aritmetico.jpg",gray_aritmentica)
```

```python
write_and_show("gray_ponderado.jpg",gray_ponderada)
```
