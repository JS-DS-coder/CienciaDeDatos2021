# Paginas totales

Ir a traer todas las paginas de YELP

```python
initText = '<span class=" css-bq71j2">'
endText = ' reviews</span>'
patron = initText + '(.+?)'+ endText
TotalReviews = re.search(patron,response.text).group(1)
TotalReviews = int(TotalReviews)
TotalReviews
```

```python
initText = '<span class=" css-e81eai">1 of'
endText = ' </span>'
patron = initText + '[0-9+] of (.+?)' + endText 
#totalPaginas = re.findall(patron,response.text).group(1)
totalPaginas = 38
```

```python
import math
totalperpage = math.floor(TotalReviews/totalPaginas)
totalperpage
```

```python
allReviews = []
initText = '<span class=" raw__373c0__tQAx6" lang="en">'
endText  = '</span>'
patron = initText + '(.+?)' + endText
for pageNum in range (0,TotalReviews,totalperpage):
  URL = 'https://www.yelp.com/biz/contramar-ciudad-de-m%C3%A9xico?osq=contramar&start='+str(pageNum)
  response = requests.get (URL)
  try:
    indexList  = [(review.start(0),review.end(0)) for review in re.finditer(patron,response.text)] 
    for index  in indexList:
      review = response.text[index[0] + len(initText): index[1]-len(endText)] ## se le suma el len de init text para quitar el span y le quito el end span
      allReviews.append(review)
  except:
      pass    
```

```python
len(allReviews)
```

```python
allReviews
```

```python
for index, review in enumerate(allReviews):
  review = review.replace('<br>','')
  review = review.replace('&#39;',"'")
  review = review.replace(u'\xa0'," ")
  review = review.lower()
  review = review.rstrip('\n')
  allReviews[index] = review
```

```python
from textblob import TextBlob

polarities = []
subjects = []
label = []
objetividad = []
review1 = []

for review in allReviews:
  sentimiento = TextBlob(review)
  review1.append(review)
  polarities.append(sentimiento.polarity)
  subjects.append(sentimiento.subjectivity)


  if (sentimiento.polarity > 0):
    label.append('positivo')
  elif (sentimiento.polarity < 0):
    label.append('negativo')
  else:
    label.append('neutral')

  if (sentimiento.subjectivity >= 0.5):
    objetividad.append('NO')
  else:  
    objetividad.append('SI')
```

```python
import pandas as pd
print(len(polarities))
print(len(subjects))
print(len(label))
print(len(objetividad))
dataOut = pd.DataFrame({'polaridad':polarities,'subjetividad': subjects,'label':label,'objetividad':objetividad,'review':review1})
```
