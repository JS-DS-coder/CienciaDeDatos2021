```python
import requests
import re
from   bs4 import BeautifulSoup
```

# Web scrapping

```python
response = requests.get('https://www.yelp.com/biz/contramar-ciudad-de-m%C3%A9xico?osq=contramar')
```

```python
response.status_code  # codigo 200 la consulta fue exitosa
```

```python
response.text
```

```python
#encontrar bajo que clase por ejemplo se encuentran los comentarios.
initText = '<span class=" raw__373c0__tQAx6" lang="en">'
endText  = '</span>'
patron = initText + '(.+?)' + endText
```

```python
review = re.search(patron,response.text).group(1)
review
```

```python
indexList  = [(review.start(0),review.end(0)) for review in re.finditer(patron,response.text)]
indexList
```

```python
for index  in indexList:
  review = response.text[index[0] + len(initText): index[1]-len(endText)] ## se le suma el len de init text para quitar el span y le quito el end span
  print (review)
  print ('----')
```
